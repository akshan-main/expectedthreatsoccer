```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("europaLeague_final.csv")
```

```python
df = df.loc[(df['TeamName']=='Manchester United') & (df['JerseyNO']==10) & (df['Action']=='Pass') & (df['Outcome']=='Successful')]
```

```python
df.head()
```

```python
xT = pd.read_csv("expected_threat.csv", header=None)
```

```python
xT
```

```python
xT = np.array(xT)
```

```python
xT_rows, xT_cols = xT.shape
```

```python
print(xT_rows,xT_cols)
```

```python
df['x1'] = pd.cut(df['x'], bins=xT_cols, labels=False)
```

```python
df['y1'] = pd.cut(df['y'], bins=xT_rows, labels=False)
```

```python
df['x2'] = pd.cut(df['endX'], bins=xT_cols, labels=False)
```

```python
df['y2'] = pd.cut(df['endY'], bins=xT_rows, labels=False)
```

```python
df.head()
```

```python
df['start_zone'] = df[['x1', 'y1']].apply(lambda x: xT[x[1]][x[0]], axis=1)
```

```python
df['end_zone'] = df[['x2', 'y2']].apply(lambda x: xT[x[1]][x[0]], axis=1)
```

# New Section

```python
df.head()
```

```python
df['xT'] = df['end_zone'] - df['start_zone']
```

```python
df.head()
```

```python
df.xT.sum()
```
