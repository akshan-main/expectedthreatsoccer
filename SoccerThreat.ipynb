```python
import pandas as pd
import numpy as np

df = pd.read_csv("europaLeague_final.csv")
```

```python
df = df.loc[(df['teamId']=='Manchester United') & (df['playerId']==10) & (df['type']=='Pass') & (df['outcome']=='Successful')]
```

```python
df.head()
```

```python
xT = pd.read_csv("expected_threat.csv", header=None)
```

```python
xT
```

```python
xT = np.array(xT)
xT_rows, xT_cols = xT.shape
```

```python
print(xT_rows,xT_cols)
```

```python
df['x1_bin'] = pd.cut(df['x'], bins=xT_cols, labels=False)
df['y1_bin'] = pd.cut(df['y'], bins=xT_rows, labels=False)
df['x2_bin'] = pd.cut(df['endX'], bins=xT_cols, labels=False)
df['y2_bin'] = pd.cut(df['endY'], bins=xT_rows, labels=False)
```

```python
df.head()
```

```python
df['start_zone_value'] = df[['x1_bin', 'y1_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
df['end_zone_value'] = df[['x2_bin', 'y2_bin']].apply(lambda x: xT[x[1]][x[0]], axis=1)
```

```python
df.head()
```

```python
df['xT'] = df['end_zone_value'] - df['start_zone_value']
```

```python
df.head()
```

```python
df.xT.sum()
```

As you can notice, playerId #10, Marcus rashford was detrimental to the team. The result suggests that Rashford made multiple wrong decisions, inhibiting Manchester United from scoring more goals in the Europa League Final.
